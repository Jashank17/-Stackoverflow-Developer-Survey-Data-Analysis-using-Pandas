## 1)Find the average age of developers when they wrote their first line of code.
### Assumptions:Younger than 5 years = 5 Years
### Older than 85 years = 85 years


```python
a=data["Age1stCode"]
Alldata=[]
for i in a:
    f=str(i)
    nu=f.isnumeric()
    if nu is True:
        Alldata.append(i)
    elif i=="Younger than 5 years":
        Alldata.append(5)
    elif i=="Older than 85 years":
        Alldata.append(85)
    else:
        Alldata.append(0)
      
Alldata = [int(i) for i in Alldata]
def average(li):
    return sum(li)/len(li)

print("Average Age of Developers when they wrote their first line of code is", average(Alldata))
```

## 4) Based on this survey, what will be the most desired programming language for the 
## year 2020

```python
desiredLanguages = []
for index, row in data["LanguageDesireNextYear"].fillna("").items():
    desiredLanguages.extend(row.split(";"))
def most_common(List): 
    return(mode(List)) 
    

print("Based on the survey, The most desired programming language for the year 2020 is ",most_common(desiredLanguages)) 
```

## 5) What is the distribution of people who code as a hobby based on gender and 
##     continent?


```python
def Gender(li):
    if li=="Man":
        return "Man"
    elif li=="Woman":
        return "Woman"
    else:
        return "Others"
```

```python
data["Gender"] = data["Gender"].apply(lambda x: Gender(x))
data["Coding Hobby"] = data["Hobbyist"].apply(lambda x: 1 if x=="Yes" else 0)
CodingHobby = data.groupby(by=["Continent","Gender"]).agg({"isHobbyist":["sum"]})
CodingHobby.columns = ["Coding Hobby"]
```
